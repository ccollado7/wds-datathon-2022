# 1. Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

#model
from catboost import CatBoostRegressor

#train/test split
from sklearn.model_selection import train_test_split

#pool class
from catboost import Pool
```

# 2. Data

```python
#paths to the Github repo

train_path = 'https://raw.githubusercontent.com/ccollado7/wds-datathon-2022/main/data/raw/train.csv'
test_path = 'https://raw.githubusercontent.com/ccollado7/wds-datathon-2022/main/data/raw/test.csv'
submit_path = 'https://raw.githubusercontent.com/ccollado7/wds-datathon-2022/main/data/raw/sample_solution.csv'
```

```python
#train dataset

train = pd.read_csv(train_path)
train.head()
```

```python
#test dataset

test = pd.read_csv(test_path)
test.head()
```

```python
#sumbit

submit = pd.read_csv(submit_path)
submit.head()
```

```python
x = train.drop(['id','site_eui'], axis=1)
x.columns
```

```python
y = train['site_eui']
```

```python
cat_features = list(train.select_dtypes(include=['object']).columns)
cat_features
```

```python
#train / validation split
x_train, x_validation, y_train, y_validation = train_test_split(x, y, test_size=0.2, random_state=0)

#train pool
train_pool = Pool(
    data=x_train, 
    label=y_train, 
    cat_features=cat_features
)

#validation pool
validation_pool = Pool(
    data=x_validation, 
    label=y_validation, 
    cat_features=cat_features
)
```

### Model 2

```python
model_2 = CatBoostRegressor(
    iterations=10000, #high number iterations
    learning_rate=0.03, #low learning rate
    early_stopping_rounds=100
)

model_2.fit(
    train_pool,
    eval_set=validation_pool,
    verbose=False,
    plot=True
);
```

```python
print(model_2.tree_count_)
```

```python
x_test = test.drop('id', axis=1)
x_test.columns
```

```python
submit['site_eui'] = model_2.predict(x_test)
submit.to_csv('modelo_2.csv',index=False)
```

### Model 3

```python
#learning rate free for the model

model_3 = CatBoostRegressor(
    iterations=10000, #high number iterations
    early_stopping_rounds=100
)

model_3.fit(
    train_pool,
    eval_set=validation_pool,
    verbose=False,
    plot=True
);
```

```python
print(model_3.tree_count_)
```

```python
submit['site_eui'] = model_3.predict(x_test)
submit.to_csv('modelo_3.csv',index=False)
```
