# 1. Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

#model
from sklearn import linear_model

#Imputer
from sklearn.impute import SimpleImputer

#Scaler
from sklearn.preprocessing import StandardScaler
```

# 2. Data

```python
#paths to the Github repo

train_path = 'https://raw.githubusercontent.com/ccollado7/wds-datathon-2022/main/data/raw/train.csv'
test_path = 'https://raw.githubusercontent.com/ccollado7/wds-datathon-2022/main/data/raw/test.csv'
submit_path = 'https://raw.githubusercontent.com/ccollado7/wds-datathon-2022/main/data/raw/sample_solution.csv'
```

```python
#train dataset

train = pd.read_csv(train_path)
train.head()
```

```python
#test dataset

test = pd.read_csv(test_path)
test.head()
```

```python
#sumbit

submit = pd.read_csv(submit_path)
submit.head()
```

```python
x_train = train.drop(['id','site_eui'], axis=1)
x_train.columns
```

```python
y_train = train['site_eui']
```

```python
x_test = test.drop('id', axis=1)
x_test.columns
```

```python
columns = ['energy_star_rating','floor_area','year_built','ELEVATION']
```

```python
x_train_sub = x_train[columns]
x_train_sub.isna().sum()
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
x_train_sub = imp.fit_transform(x_train_sub)
```

```python
scaler = StandardScaler()
x_train_sub = scaler.fit_transform(x_train_sub)
```

```python
x_train_sub
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(x_train_sub,y_train)
```

```python
x_test_sub = x_test[columns]
x_test_sub.isna().sum()
```

```python
x_test_sub = imp.transform(x_test_sub)
x_test_sub = scaler.transform(x_test_sub)
```

```python
submit['site_eui'] = reg.predict(x_test_sub)
submit.to_csv('modelo_lineal_1.csv',index=False)
```
